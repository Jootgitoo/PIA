# INGRESOS MENSUALES

# 1. Generamos un conjunto de datos

```python
import numpy as np

#Generamos los años de experiencia
    #Generamos 100 valores aleatorios entre 0 y 1
    #Luego se multiplican por 20 por eso los años son de 0 a 20.
X = 20 * np.random.rand(100, 1) 
X

#Creamos el ruido
ruido = 5 * np.random.randn(100, 1)

#Generamos los ingresos
y = 30 + (4 * X) + ruido
y
```

# 2. Ajustamos el modelo de regresion lineal

## 2.1 Entrenamos al modelo

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X, y)
```

# 3. Mostramos la gráfica con los datos 

```python
import matplotlib.pyplot as plt
plt.plot(X, y, "b.")
plt.xlabel("Experiencia (años)")
plt.ylabel("Sueldo (Miles)")
plt.show()
```

# 4. Creamos el DataFrame

```python
import pandas as pd
df = {
    "experiencia_laboral": X,
    "sueldo_en_miles": y
}
df
```

# 5. Visualizamos los resultados con la línea de regresion

```python
# Predecimos el valor minimo y el valor máximo
X_min_max = np.array([[df["experiencia_laboral"].min()], [df["experiencia_laboral"].max()]])
y_train_pred = linear_regression.predict(X_min_max)

# Representación gráfica de la función hipótesis generada
plt.plot(X_min_max, y_train_pred, "g-")    
plt.plot(df["experiencia_laboral"], df["sueldo_en_miles"], "b.")
plt.xlabel("experiencia_laboral")
plt.ylabel("sueldo_en_miles")
plt.show()
```
