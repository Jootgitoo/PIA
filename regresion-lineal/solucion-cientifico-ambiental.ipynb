# CIENTÍFICO AMBIENTAL

# 1. Generamos los datos sintéticos

```python
# Generamos la distancia (metros)
import numpy as np

#Generamos distancias entre 50 y 500 metros
X = np.random.uniform(50, 500, (1000, 1))

#Generamos el ruido
ruido = 5 * np.random.randn(1000, 1)

# Generamos la concentración de contaminantes (siguiendo una fórmula y = a/X + b + ruido (a y b son constantes inventadas))
y = 200 / X + 10 + ruido
```

# 2. Ajustamos el modelo

### Estimamos las concentraciones de contaminantes en función de la distancia

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X, y)
```

```python
import matplotlib.pyplot as plt
plt.plot(X, y, "b.")
plt.xlabel("Distancia")
plt.ylabel("Contaminación")
plt.show()
```

# 3. Visualizamos los datos

## 3.1 Creamos el DataFrame

```python
import pandas as pd
df = {
    "distancia": X,
    "contaminacion": y
}
```

# 3.2 Mostramos los resultados

```python
# Predecimos el valor minimo y el valor máximo
X_min_max = np.array([[df["distancia"].min()], [df["distancia"].max()]])
y_train_pred = linear_regression.predict(X_min_max)

# Representación gráfica de la función hipótesis generada
plt.plot(X_min_max, y_train_pred, "g-")    
plt.plot(df["distancia"], df["contaminacion"], "b.")
plt.xlabel("Distancia")
plt.ylabel("Contaminación")
plt.show()
```
