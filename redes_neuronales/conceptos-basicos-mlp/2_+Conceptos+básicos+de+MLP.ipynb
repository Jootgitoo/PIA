<a id="section1"></a>
# <font color="#004D7F" size=6>1. Redes Neuronales en PyTorch</font>

- PyTorch ofrece una amplia gama de funcionalidades, pero su aplicación más frecuente es la creación de modelos de aprendizaje profundo.
- La estructura más básica se puede diseñar empleando la clase `Sequential`, que consiste en una serie de capas interconectadas de manera lineal.
- Es posible generar un modelo `Sequential` y establecer todas sus capas en un solo paso.
- Es necesario especificar todas las capas dentro de los paréntesis, siguiendo el orden de procesamiento desde la entrada hasta la salida.

A modo de ilustración:

````python
pip install torch
````

````python
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(764, 100), #Input de 764 y capa oculta de 100
    nn.ReLU(), # Función de activación
    nn.Linear(100, 50), #La entrada será la salida de la anterior 100 y la siguiente capa de 50
    nn.ReLU(), # Función de activación
    nn.Linear(50, 10), #entrada de 50 y salida de 10
    nn.Softmax()
)
print(model)
````

Otra forma de usar `Sequential` es pasar un diccionario ordenado en el que puedes asignar nombres a cada capa:

````python
from collections import OrderedDict
import torch.nn as nn

model = nn.Sequential(OrderedDict([
    ('dense1', nn.Linear(764, 100)),
    ('act1', nn.ReLU()),
    ('dense2', nn.Linear(100, 50)),
    ('act2', nn.ReLU()),
    ('dense3', nn.Linear(50, 10)),
    ('out', nn.Softmax())
]))
print(model)
````

Y si deseas construir las capas una por una en lugar de hacerlo todo de una vez, puedes hacer lo siguiente:

````python
import torch
import torch.nn as nn

model = nn.Sequential()
model.add_module('dense1', nn.Linear(764, 100))
model.add_module('act1', nn.ReLU())
model.add_module('dense2', nn.Linear(100, 50))
model.add_module('act2', nn.ReLU())
model.add_module('dense3', nn.Linear(50, 10))
model.add_module('out', nn.Softmax())
print(model)
````

# <font color="#004D7F" size=6>2. Entradas del Modelo</font>

- La capa inicial de tu modelo determina la estructura de los datos de entrada.
- En la ilustración previa, se observa `nn.Linear(764, 100)` como primera capa.
- Los parámetros de esta capa pueden variar según su tipo, pero en este caso específico, al tratarse de una capa lineal (también denominada capa densa o completamente conectada), los dos argumentos representan las dimensiones de __entrada y salida__ respectivamente.

Es importante destacar que el tamaño del lote no se especifica explícitamente. En este ejemplo concreto, se espera que se proporcione a esta capa un tensor de PyTorch con forma `(n, 764)`, y se obtendrá como resultado un tensor de forma `(n, 100)`, donde `n` corresponde al tamaño del lote.

# <font color="#004D7F" size=6>3. Capas, Activaciones y Propiedades de la Capa</font>

PyTorch ofrece una amplia variedad de capas para redes neuronales, e incluso permite la creación de capas personalizadas. Algunas de las capas más frecuentes son:
- `nn.Linear(input, output)`: Capa totalmente conectada.
- `nn.Conv2d(in_channel, out_channel, kernel_size)`: Capa de convolución 2D, muy utilizada en redes para procesamiento de imágenes.
- `nn.Dropout(probability)`: Capa de dropout, empleada para introducir regularización. (Para evitar overfitting)
- `nn.Flatten()`: Transforma un tensor de entrada multidimensional en una dimensión (por cada muestra en un lote).

Además de las capas, existen funciones de activación que se aplican elemento a elemento en un tensor. Estas funciones suelen utilizarse después de una capa y antes de la siguiente. Algunas funciones de activación comunes son:

- `nn.ReLU()`: Unidad lineal rectificada, la más utilizada actualmente.
- `nn.Sigmoid()` y `nn.Tanh()`: Funciones sigmoidea y tangente hiperbólica, populares en modelos más antiguos.
- `nn.Softmax()`: Convierte un vector en valores similares a probabilidades, frecuente en redes de clasificación.

La mayoría de los componentes aceptan dos argumentos adicionales: el dispositivo y el tipo de datos.

### Dispositivo
Se puede elegir entre CPU y GPU, o dejar que PyTorch lo decida. Para especificar un dispositivo:

```python
nn.Linear(764, 100, device="cpu")
```

O para una GPU NVIDIA compatible:

```python
nn.Linear(764, 100, device="cuda:0")
```

### Tipo de datos
El argumento `dtype` especifica el tipo de datos con el que operará la capa. Generalmente se usa un flotante de 32 bits, pero si se necesita un tipo diferente:

```python
nn.Linear(764, 100, dtype=torch.float16)
```

# <font color="#004D7F" size=6>4. Función de Pérdida y Optimizadores de Modelo</font>

El entrenamiento de una red neuronal busca optimizar los pesos para generar la salida deseada. En el aprendizaje profundo, el método utilizado para optimizar estos pesos es el descenso de gradiente.

- Existen diversas variantes del descenso de gradiente.
- Puedes seleccionar una preparando un optimizador para tu modelo.
- Su uso implica definir una función de pérdida y minimizarla mediante el optimizador.
- La función de pérdida proporciona una puntuación de distancia que indica cuán alejada está la salida del modelo de la salida deseada.

### Función de pérdida

En PyTorch, puedes manipular directamente el tensor de salida del modelo para calcular la pérdida. Sin embargo, también puedes utilizar las funciones proporcionadas por PyTorch para este fin, por ejemplo:

```python
loss_fn = nn.CrossEntropyLoss()
loss = loss_fn(output, label)
```

En este caso, `loss_fn` es una función, y `loss` es un tensor que admite diferenciación automática. Puedes iniciar la diferenciación llamando a `loss.backward()`.

Algunas funciones de pérdida comunes en PyTorch son:

- `nn.MSELoss()`: Error cuadrático medio, útil en problemas de regresión.
- `nn.CrossEntropyLoss()`: Pérdida de entropía cruzada, útil en problemas de clasificación.
- `nn.BCELoss()`: Pérdida de entropía cruzada binaria, útil en problemas de clasificación binaria.

<i class="fa fa-info-circle" aria-hidden="true"></i>

### Optimizador

La creación de un optimizador es similar:

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

Algunos optimizadores comunes son:

- `torch.optim.Adam()`: El algoritmo Adam (estimación de momento adaptativo).
- `torch.optim.NAdam()`: El algoritmo Adam con momento
- `torch.optim.SGD()`: Descenso de gradiente estocástico
- `torch.optim.RMSprop()`: El algoritmo RMSprop

# <font color="#004D7F" size=6>5. Entrenamiento e Inferencia de Modelos</font>

PyTorch no tiene una función dedicada para el entrenamiento y la evaluación de modelos.
- Un modelo definido por sí mismo es como una función.
- Pasas un tensor de entrada y obtienes de vuelta un tensor de salida.
- Por lo tanto, es tu responsabilidad escribir el bucle de entrenamiento.





Un bucle de entrenamiento mínimo es como el siguiente:

```python
for n in range(num_epochs):
    y_pred = model(X)
    loss = loss_fn(y_pred, y)
    optimizer.zero_grad() #Para evitar acumulación de gradientes de pasos anteriores
    loss.backward() #Calculamos pesos pero no los cambiamos
    optimizer.step() #Cambia los pesos
```

Si ya tienes un modelo, puedes simplemente tomar `y_pred = model(X)` y usar el tensor de salida `y_pred` para otros propósitos. Así es como utilizas el modelo para predicciones o inferencia.

# <font color="#004D7F" size=6>6. Examinación de un Modelo</font>

Una vez que tienes un modelo, puedes comprobar qué es imprimiéndolo:

````python
print(model)
````

Si deseas guardar el modelo, puedes usar la biblioteca `pickle` de Python. Pero también puedes acceder a él utilizando PyTorch:

````python
torch.save(model, "modelo_guardado.pth")
````



De esta manera, tienes todo el objeto del modelo guardado. Puedes recuperar el modelo con:

````python
model1 = torch.load("modelo_guardado.pth", weights_only=False)
model1
````

Pero la forma recomendada de guardar un modelo es dejar el diseño del modelo en el código y guardar solo los pesos (lo veremos en detalle más adelante). Puedes hacerlo así:

````python
torch.save(model.state_dict(), "modelo_guardado2.pth")
````


La función `state_dict()` extrae solo los estados (es decir, los pesos en un modelo). Para recuperarlo, necesitas reconstruir el modelo desde cero y luego cargar los pesos de la siguiente manera:

````python
from collections import OrderedDict

model = nn.Sequential(OrderedDict([
    ('dense1', nn.Linear(764, 100)),
    ('act1', nn.ReLU()),
    ('dense2', nn.Linear(100, 50)),
    ('act2', nn.ReLU()),
    ('dense3', nn.Linear(50, 10)),
    ('out', nn.Sigmoid())
]))
model.load_state_dict(torch.load("modelo_guardado2.pth"))
````
