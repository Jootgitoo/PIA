# <font color="#004D7F" size=6> 0. Contexto</font>

Las redes neuronales se construyen con capas conectadas entre sí. Existen diferentes tipos de capas. Para aplicaciones relacionadas con imágenes, siempre encontrarás capas convolucionales. Estas capas tienen muy pocos parámetros, pero se aplican sobre una entrada de gran tamaño. Son poderosas porque pueden preservar la estructura espacial de la imagen. Por lo tanto, se utilizan para producir resultados en visión por computadora. Después de completar esta clase, sabrás:

- Qué son las capas convolucionales y las capas de pooling
- Cómo encajan en una red neuronal
- Cómo se diseña una red neuronal que usa capas convolucionales

# <font color="#004D7F" size=6>1. Redes neuronales convolucionales</font>

- Ejemplo: Redes neuronales para procesar imágenes en escala de grises:

  - Caso más simple en visión por computadora con aprendizaje profundo

- Características de una imagen en escala de grises o RGB:

  - Arreglo de píxeles

  - Cada píxel: valor entre 0 y 255
  
  - Ejemplo: imagen 32x32 = 1024 píxeles

- Desafíos para redes neuronales tradicionales:

  - Primera capa requeriría al menos 1024 pesos de entrada

  - Dificultad para extraer información útil de valores de píxeles individuales

  - Importancia de la estructura espacial (ej. líneas horizontales o verticales)

- Ventajas de las redes neuronales convolucionales (CNN):

  - Uso de capas convolucionales para preservar información espacial

  - Aprendizaje de relaciones entre píxeles vecinos

  - Generación de representaciones de características

- Características clave de las CNN:
  - Invariancia a ciertas distorsiones (color, rotación, escala)

  - Menor número de pesos, facilitando el entrenamiento

- Beneficios de las CNN para procesamiento de imágenes:
  - Mejor capacidad para extraer información relevante
  
  - Mayor robustez frente a variaciones en la imagen de entrada

# <font color="#004D7F" size=6>2. Arquitectura de las Redes Neuronales Convolucionales</font>

Esta estructura permite a las CNN procesar eficientemente imágenes, extrayendo características relevantes y realizando tareas como clasificación de manera efectiva.

- Tipos de capas en una red neuronal convolucional simple:
  1. Capas convolucionales

  2. Capas de pooling

  3. Capas completamente conectadas

- **Capa de convolución**:
  - Neuronas llamadas "filtros"

  - Usualmente convolución 2D para imágenes

  - Filtro: parche 2D (ej. 3x3 píxeles)

  - Campo receptivo: tamaño del parche

  - Proceso: multiplicación y suma de píxeles

  - Múltiples filtros generan mapas de características

  - Salida: tensor de mapas de características apilados

- Función de los mapas de características:
  - Aprenden características de la imagen

  - Ejemplo: detección de líneas verticales

  - Capas apiladas infieren características de alto nivel

- **Capa de Agrupación (Pooling)**:
  - Reduce el muestreo de mapas de características

  - Consolida características aprendidas

  - Comprime y generaliza representaciones

  - Tipos: average pooling, max pooling

- **Capa totalmente conectada**:
  - Generalmente en las capas finales

  - Procesa características consolidadas

  - Produce la predicción final

  - Puede haber varias capas apiladas

  - En clasificación: última capa seguida de función softmax
