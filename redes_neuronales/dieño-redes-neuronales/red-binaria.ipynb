### Imports necesarios para la actividad

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
```

### Cargamos el archivo

En nuestro caso el dataset tiene solo 2 columnas, la salida hamm o spam y el mensaje

```python
import numpy as np

#utilizamos el comando skiprows para que el modelo no coja la primera fila que es la cabezera
dataset = np.loadtxt('Datasets/lie_dataset_100.csv', delimiter=',', skiprows=1)
X = dataset[:, 0:4]
y = dataset[:, 4]
print(X[0:5])
print(y[0:5])
```

### Tranformación de los datos

Tenemos que transformar los datos ya que PyTorch y NumPy utilizan puntos flotantes diferentes

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y.reshape(-1, 1), dtype=torch.float32)

print(X[0:5])
print(y[0:5])
```

### Definimos el modelo

```python
model = nn.Sequential(
    nn.Linear(4, 8),    # 4 features de entrada, capa oculta de 8
    nn.ReLU(),
    nn.Linear(8, 4),    # reducimos a 4
    nn.ReLU(),
    nn.Linear(4, 1),    # salida binaria (lie o no lie)
    nn.Sigmoid()
)
print(model)
```

### Funcion de perdida

```python
loss_fn = nn.BCELoss() #al ser clasificación binaria establecemos binary cross entropy
optimizer = optim.Adam(model.parameters(), lr=0.001) #Necesita saber qué parámetros optimizar
```

### Entenamos el modelo

```python
n_epochs = 500
batch_size = 10
num_samples = min(len(X), len(y)) # Use the minimum length to ensure X and y are always aligned

for epoch in range(n_epochs):
  for i in range(0, num_samples, batch_size):
    # Define the end index for the current batch, ensuring it doesn't exceed num_samples
    current_batch_end = min(i + batch_size, num_samples)
    Xbatch = X[i:current_batch_end] #Partición del conjunto de datos
    ybatch = y[i:current_batch_end] #Corresponding partition for labels

    # Skip if the batch is empty (e.g., if num_samples was 0 or due to an edge case)
    if Xbatch.shape[0] == 0:
      continue

    #Forward
    y_pred = model(Xbatch)
    loss = loss_fn(y_pred, ybatch) # Calculo la pérdida
    optimizer.zero_grad()
    loss.backward()
    optimizer.step() # Actualiza pesos
  print(f'Finished epoch {epoch}, latest loss {loss}')
```

### Evaluamos el modelo

```python
with torch.no_grad(): # para que no tenga en cuenta derivaciones anteriores
  y_pred = model(X)
  accuracy = (y_pred.round() == y).float().mean()
  print(f"Accuracy {accuracy}")
```

```python
# Sin no grad
y_pred = model(X)
accuracy = (y_pred.round() == y).float().mean()
print(f"Accuracy {accuracy}")
```
