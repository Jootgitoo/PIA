# <font color="#004D7F" size=6>1. Creando tensores</font>

Si estás familiarizado con NumPy, recordarás que hay varias formas de crear una matriz. Lo mismo ocurre en PyTorch para crear tensores. La forma más sencilla de crear un tensor constante específico es la siguiente:

```python
pip install torch
```

```python
import torch
a = torch.tensor([[1, 2, 4], [4, 5, 6]], dtype=torch.int32)
print(a)
```

El argumento `dtype` especifica el tipo de datos de los valores en el tensor y es opcional. También puedes convertir una matriz de NumPy en un tensor de PyTorch fácilmente.

Por lo general, crearías un tensor para un propósito específico. Por ejemplo, si deseas tener diez valores distribuidos uniformemente entre -1 y 1, puedes usar la función `linspace()`:


```python

import torch
a = torch.linspace(-1,1,10)
print(a)
```

Tensor de valores aleatorios, lo cual es muy útil para probar tus funciones

```python
import torch
a = torch.rand(3, 4)
print(a)
```

Este tensor resultante tiene una dimensión de 3x4 y cada valor se distribuye uniformemente entre 0 y 1.

Si desea que los valores se distribuyan normalmente, simplemente cambie la función a `randn()`:

```python
a = torch.randn(3, 4)
print(a)
```

Si desea que los valores aleatorios sean un número entero, e.g., entre 3 y 10, puede utilizar la función `randint()`:

Los valores están en el rango 3≤x<10.

```python

import torch
a = torch.randint(3, 10, size=(3,4))
print(a)

```

De forma predeterminada, el límite inferior es cero, por lo que si desea que los valores sean 0≤x<10, puede usar:


```python
a = torch.randint(10, size=(3,4))
print(a)
```

Los otros tensores comúnmente utilizados son el tensor cero y los tensores con todos los valores iguales.

Para crear un tensor cero (por ejemplo, de dimensión 2×3×4), puedes usar:


```python
import torch
a = torch.zeros(2, 3, 4)
print(a)
```


Y para crear un tensor de todos los valores iguales a 5, puedes usar:


```python
import torch
a = torch.full((2, 3, 4), 5)
print(a)
```

Pero si quieres que todos los valores sean 1, existe una función más sencilla:

```python
a = torch.ones(2, 3, 4)
print(a)
```

Finalmente, si quieres una matriz de identidad, puedes obtenerla con `diag()` o `eye()`:

```python
a = torch.diag(torch.randn(4))
print(a)
print()
a = torch.eye(4)
print(a)
```

# <font color="#004D7F" size=6>2. Comprobando un tensor</font>

Una vez que tenga un tensor y quiera saber más sobre él, simplemente puede imprimirlo en la pantalla usando `print()`.

Pero si el tensor es demasiado grande, es más fácil mostrar su dimensión comprobando su forma con `shape` o `size`:


```python
import torch
a = torch.zeros(2, 3, 4)
print(a.shape)
print(a.size())
```

Si quieres ver cuántas dimensiones tienes (i.e., 2×3×4 es 3 y 3×4 es 2), puedes leer la propiedad `ndim`:

```python
print(a.ndim)
```

Si usa `len()` para verificar un tensor, solo le dará el tamaño de la primera dimensión, por ejemplo:

```python
import torch
a = torch.zeros(2, 3, 4)
print(len(a)) #Tamaño de la primera dimensión
print(len(a[0]))
print(len(a[0][0]))
```

Otra propiedad el tipo de datos con la propiedad `dtype`:

```python
print(a.dtype)
```


Si desea cambiar el tipo de datos, puede recrear el tensor con un nuevo tipo:


```python

import torch
a = torch.zeros(2, 3, 4)
b = a.type(torch.int32)
print(a.dtype)
print(b.dtype)
```

# <font color="#004D7F" size=6>3. Manipulación de tensores</font>

Una operación común es cambiar la forma del tensor. Por ejemplo, es posible que desees convertir un tensor 2D en 1D.

```python
a = torch.randn(3, 4, 5)
print(a)
print(a[1])
```

Cortar un tensor en dos dimensiones

```python
print(a[1:, 2:4])
```

También puede utilizar la misma sintaxis de corte para agregar una nueva dimensión.
`None` se utiliza insertar una nueva dimensión en un lugar específico.

```python
print(a.shape)
print(a[:, None, :, None].shape)
```

En NumPy existe una función `expand_dims()` para este propósito, pero PyTorch no la proporciona. Una función similar es `unsqueeze()`:

```python
b =torch.unsqueeze(a, dim=2)
print(a.shape)
print(b.shape)
print(a)
print(b)
```


Para convertir un tensor 2D en 1D, puedes usar:

```python
import torch
a = torch.randn(3, 4)
print(a)
print(a.ravel()) #Lo convertimos a un vector
```

También puedes usar la función `reshape()`:

```python
print(a.reshape(-1))
```

LA transposición matricial

```python
print(a)
print()
print(a.t())
```


Pero la función `transpose()` en PyTorch, a diferencia de NumPy, requiere que especifiques explícitamente los dos ejes para intercambiar:

```python
print(a.transpose(0, 1))
```

Si tiene varios tensores, puede combinarlos apilándolos CON `vstack()` para verticalmente y `hstack()` para horizontalmente.

```python
a = torch.randn(3, 3)
b = torch.randn(3, 3)
print(a)
print()
print(b)
print()
print(torch.vstack([a, b]))
print()
print(torch.hstack([a,b]))
```


La función `concat()` es similar a `vstack()`:

```python
c = torch.concat([a,b])
print(c)
```

Lo contrario es dividir, por ejemplo:

```python
print(torch.vsplit(c, 2))
```

# <font color="#004D7F" size=6>4. Funciones tensoriales</font>

Los tensores de PyTorch se pueden tratar como matrices. Por lo tanto, a menudo puedes usarlos de manera similar a las matrices NumPy. Por ejemplo, tienes las funciones de funciones matemáticas comunes:

```python
a = torch.randn(2, 3)
print(a)
print("exp =", torch.exp(a))
print()
print("log =", torch.log(a))
print()
print("sin =", torch.sin(a))
print()
print("arctan =", torch.atan(a))
print()
print("abs =", torch.abs(a))
print()
print("cuadrado =", torch.square(a))
print()
print("sqrt =", torch.sqrt(a))
print()
print("ceil =", torch.ceil(a))
print()
print("round =", torch.round(a))
print()
# Ajusta todos los valores menores que 0 a 0 y todos los valores mayores que 1 a 1.
print("clip =", torch.clamp(a, 0, 1))
```


Tenga en cuenta que si una función no está definida (por ejemplo, raíz cuadrada de números negativos), el resultado será `nan`, pero no se generará ninguna excepción. En PyTorch, tienes una función para comprobar si los valores de un tensor son `nan`:

```python
b = torch.sqrt(a)
print(b)
print(torch.isnan(b))
```


De hecho, además de estas funciones definidas, los operadores de Python también se pueden aplicar a los tensores:

```python
a = torch.randn(2, 3)
b = torch.randn(2, 3)
print(a)
print()
print(b)
print()
print(a+b)
print()
print(a/b)
print()
# Exponente
print(a**2)
```

Multiplicar matrices:

```python
print(torch.matmul(a, b.T))
print()
print(a @ b.T)
```

El operador `@` de Python también se puede utilizar para productos escalares vectoriales

```python
a = torch.randn(3)
b = torch.randn(3)
print(a)
print()
print(b)
print()
print(torch.dot(a,b))
print()
print(a @ b)
```


Si tratas los valores de un tensor como muestras, es posible que también desees encontrar algunas estadísticas al respecto. Algunas también se proporcionan en PyTorch:

```python
a = torch.randn(3, 4)
print(a)
print()
print(torch.mean(a, dim=0))
print()
print(torch.std(a, dim=0))
print()
print(torch.cumsum(a, dim=0)) #Suma acumulada por columna de la dimensión 0
print()
print(torch.cumprod(a, dim=0)) #Producto acumulado por columna de la dimensión 0
```

Para las funciones de álgebra lineal utilizamos el submódulo `linalg`:

```python
print(torch.linalg.svd(a))
```

Y específicamente para CNN, el relleno de un tensor.

Este ejemplo de la función `pad()` es crear un relleno `(1, 1)` en la dimensión 0 y `(0, 2)` en la dimensión 1. En otras palabras, para cada dimensión 0 (filas), agregamos un valor ficticio (0) al principio y al final. Para cada dimensión 1 (columnas), agregamos cero valores ficticios al principio pero dos valores ficticios al final.

```python
b = torch.nn.functional.pad(a, (1, 1, 0, 2), value=0) #1 lado izquierdo, 1 lado derecho, 0 filas arriba, 2 filas abajo
print(a)
print()
print(b)
```

Finalmente, dado que los tensores de PyTorch pueden considerarse matrices, puedes usarlos directamente con otras herramientas como `Matplotlib`. A continuación se muestra un ejemplo de cómo trazar una superficie usando tensores de PyTorch:

```python
import matplotlib.pyplot as plt
# create tensors
x = torch.linspace(-1, 1, 100)
y = torch.linspace(-2, 2, 100)
# create the surface
xx, yy = torch.meshgrid(x, y, indexing="xy") # xy-indexing is matching numpy
z = torch.sqrt(1 - xx**2 - (yy/2)**2)
print(xx)
```


Qué valores son idénticos a lo largo de la segunda dimensión. Lo mismo ocurre con el tensor `yy`, excepto que la repetición está en la otra dimensión. La trama creada es:

Este código hace esto:

Crea una cuadrícula (x, y)
Calcula una altura z usando una ecuación matemática
Forma una superficie 3D tipo:
👉 “media esfera estirada en el eje y”
La dibuja en 3D

```python

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection="3d")
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([0, 2])
ax.plot_surface(xx, yy, z, cmap="cividis")
ax.view_init(45, 35)
plt.show()
```

```python
# create tensors
x = torch.linspace(-5, 5, 100)
y = torch.linspace(-5, 5, 100)
# create the surface
xx, yy = torch.meshgrid(x, y)
z = torch.sin(xx**2 + yy**2)

plt.imshow(z, extent=(-5, 5, -5, 5), origin='lower')
plt.colorbar()
plt.show()
```
